```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = [2099973, 1148669, 987977]
y  = [data[0]/y for y in data]
x = ['Scalar', 'Vector', 'Highway']

plt.bar(x, y)
plt.title('Speedup of Highway FMA over Scalar and Vector NO -march=native')
plt.savefig('FMASpeedup.svg')
```
